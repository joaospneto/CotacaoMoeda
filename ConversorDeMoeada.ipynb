```python
import requests

import json

cotacoes = requests.get('https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL')
cotacoes_dic = cotacoes.json()

print('A cotação atual do {} é {}.'.format(cotacoes_dic['USDBRL']['name'], cotacoes_dic['USDBRL']['bid']))
print('A cotação atual do {} é {}.'.format(cotacoes_dic['EURBRL']['name'], cotacoes_dic['EURBRL']['bid']))
print('A cotação atual do {} é {}.'.format(cotacoes_dic['BTCBRL']['name'], cotacoes_dic['BTCBRL']['bid']))
```

```python
# Consumindo as informações da API para puxar a cotação dos últimos 100 dias
cotacoes_dias = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/100')
cotacoes_dias_dic = cotacoes_dias.json()

# Transformar o dicionário cotacoes_dias_dic em uma lista com as cotações do dolar nos últimos 100 dias
lista_cotacoes = []

for item in cotacoes_dias_dic:
    lista_cotacoes.append(float(item['bid']))
lista_cotacoes.reverse()


# Plotando o gráfico com o módulo matplotlib.pyplot
import matplotlib.pyplot as plt
plt.ylabel('Cotação do Dólar')
plt.xlabel('Dias')

plt.plot(lista_cotacoes)
plt.show()
```
